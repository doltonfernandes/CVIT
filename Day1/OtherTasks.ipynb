```
import numpy as np
import matplotlib.pyplot as plt
```

```
ind = np.arange(0,3*np.pi,0.1)
y1 = np.sin(ind)
y2 = np.cos(ind)
plt.plot(ind,y1,label='sine')
plt.plot(ind,y2,label='cosine')
plt.scatter(y2+0.1*np.random.rand(y2.size),y1+0.1*np.random.rand(y1.size),label = 'ellipse')
plt.legend()
plt.show()
```

```
w = 20 * np.array([0, 1])
plt.plot([0, 1], 10 * np.array([1, 1]), label = 'line')
plt.plot([0, w[0]], [0, w[1]], 'r-', label = 'w-vector')
plt.legend(loc = "lower right")
```

```
from math import sqrt
a = 3
b = 2
X = np.linspace(-3,3,200)
Y = np.vectorize(sqrt)(np.abs(b**2*(1-(X/a)**2)))
plt.figure()
plt.plot(X,Y)
plt.plot(X,-Y)
plt.title('Ellipse')
plt.show()
```

```
import math
def distanceFromOrigin(a, b, c):
  return abs(c) / math.sqrt(a**2 + b**2)
```

```
distanceFromOrigin(1, 0, -50)
```

```
def distanceBwTwoLines(a1, b1, c1, a2, b2, c2):
  return abs(distanceFromOrigin(a1, b1, c1) - distanceFromOrigin(a2, b2, c2))
```

```
distanceBwTwoLines(1, 0, -50, 0, 1, -25)
```

```
def sigmoid(x):
    return 1  /(1+np.exp(-x))
X = np.linspace(-20,20)
plt.plot(X,np.vectorize(sigmoid)(0.5*X))
```

```
X = np.linspace(-50, 50, num=200)
Y = ( np.exp(X) - np.exp(-X) ) / ( np.exp(X) + np.exp(-X) )
plt.plot(X, Y)
```

```
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D  

def getZ(x, y):
    return - ( a*X + b*Y + d ) / c

a = 1
b = 1
c = 3
d = 10
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = y = np.arange(-3.0, 3.0, 0.05)
X, Y = np.meshgrid(x, y)
zs = np.array(getZ(np.ravel(X), np.ravel(Y)))
Z = zs.reshape(X.shape)

ax.plot_surface(X, Y, Z)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
